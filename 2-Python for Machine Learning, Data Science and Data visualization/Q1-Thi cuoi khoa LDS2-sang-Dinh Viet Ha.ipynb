```python
Q1 = df["B_Ownership"].quantile(0.25)
Q3 = df["B_Ownership"].quantile(0.75)
IQR = Q3 - Q1
df2 = df["B_Ownership"]
IQR
```

```python
condi = (df2 < (Q1 - 1.5*IQR)) | (df2 > (Q3 + 1.5*IQR))
```

```python
df2_out = df2[~(condi)]
df2_out.shape
print("So dong con lai sau khi loai outliers: ",df2_out.shape[0])
```
