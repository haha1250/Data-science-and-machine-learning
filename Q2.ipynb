```python
df_draw = df[df["SHOOTING"] != "N"].sample(n=100)
```

```python
map1 = folium.Map(location=[42.32272473, -71.08877863], zoom_start=12)
shooting = folium.map.FeatureGroup()

for lat,long in zip(df_draw["Lat"],df_draw["Long"]):
    shooting.add_child(folium.CircleMarker([lat,long],
                  radius=5, color="yellow", fill=True, 
                   fill_color="green", fill_opacity=0.6)
                  )

latitudes = list(df_draw["Lat"])
longitudes = list(df_draw["Long"])
labels = list(df_draw["OFFENSE_DESCRIPTION"])

for lat,long,label in zip(latitudes,longitudes,labels):
    folium.Marker([lat,long], popup=label).add_to(map1)
    
map1.add_child(shooting)
```
