# Splitting and applying model

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=2)
```

```python
clf = svm.SVC(C=100,gamma=0.001, kernel='rbf')
```

```python
from datetime import datetime, timedelta
```

```python
start_time = datetime.now()
```

```python
clf.fit(X_train,y_train)
```

```python
end_time = datetime.now()
```

```python
dt = end_time - start_time
seconds_1 = (dt.days*24*60*60 + dt.seconds)
print(seconds_1)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
print("Accuracy score is: ", accuracy_score(y_test,y_pred)*100,'%')
```

```python
print("The train score is: ", clf.score(X_train,y_train)*100,'%')
print("The test score is: ", clf.score(X_test,y_test)*100,'%')
```

The model fit to 89% of the dataset. --> The model fits well. 

The train score is 10% higher than the test score --> It would be better if we could bring the two score closer while remaining the high accuracy.

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

Class 8 has the lowest precision among 10 classes. Most of them was classified into other classes.

```python
target.value_counts()
```

The dataset is balanced between samples.

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=7)
```

```python
pca.fit(data)
```

```python
data_trans = pca.transform(data)
```

```python
data_trans.shape
```

```python
data.shape
```

```python
data_new = pd.DataFrame(data_trans, columns=['Principal Component 1', 'Principal Component 2',
                                             'Principal Component 3','Principal Component 4',
                                             'Principal Component 5','Principal Component 6','Principal Component 7'])
```

```python
data_new.head()
```

```python
X_train_n, X_test_n, y_train_n, y_test_n = train_test_split(data_new, target, test_size=0.2, random_state=2)
```

```python
clf_n = svm.SVC(C=100, gamma=0.001, kernel='rbf')
```

```python
clf_n.fit(X_train_n,y_train_n)
```

```python
y_pred_n = clf_n.predict(X_test_n)
```

```python
print("The accuracy score is: ", accuracy_score(y_test_n,y_pred_n)*100,'%')
```

```python
print("The training score is: ", clf_n.score(X_train_n,y_train_n)*100,'%')
print("The testing score is: ", clf_n.score(X_test_n,y_test_n)*100,'%')
```

```python
print(confusion_matrix(y_test_n,y_pred_n))
print(classification_report(y_test_n,y_pred_n))
```

```python
s_time = datetime.now()
```

```python
e_time = datetime.now()
```

```python
dt2 = e_time - s_time
seconds_2 = (dt2.days*24*60*60 + dt2.seconds)
print(seconds_2)
```

```python
y_pred_n1 = clf_n1.predict(X_test_n1)
```

```python
print("The accuracy score is: ", accuracy_score(y_test_n1,y_pred_n1)*100,'%')
```

```python
print("The training score is: ", clf_n1.score(X_train_n1,y_train_n1)*100,'%')
print("The testing score is: ", clf_n1.score(X_test_n1,y_test_n1)*100,'%')
```

```python
print(confusion_matrix(y_test_n1,y_pred_n1))
print(classification_report(y_test_n1,y_pred_n1))
```

#### Remark

After reducing dimensions from 16 to 8 using PCA, the precision score for class 8 has increased sharply (from 50% to 80%).

Moreover, the time required to train the model is shortened (from 6s to 2s), the difference in score between train and test set is slighly smaller. Test score is higher after PCA.

Applying PCA is better for predicting classes of the dataset.
